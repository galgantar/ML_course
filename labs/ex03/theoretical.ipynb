```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[401, -201], [-800, 401]])
y = np.array([200, -200])
w = np.linalg.solve(X, y)

print(X)
print(w)
print(X @ w)
print(np.linalg.cond(X))
```

```python
w = 10
y = 20
epsilon = 1e-10

x = np.linspace(-10, 10, 100)
loss = (x * w - y)**2 / (y**2 + epsilon)

plt.plot(x, loss)
plt.show()
```
